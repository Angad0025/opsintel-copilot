```python
import os

os.makedirs("/tmp/.dbt", exist_ok=True)

profiles = """
opsintel:
  target: dev
  outputs:
    dev:
      type: databricks
      catalog: workspace
      schema: opsintel_copilot
      host: dbc-f7dc29ce-1820.cloud.databricks.com
      http_path: /sql/1.0/warehouses/dbcd2a56cffd0a5e
      token: YOUR_TOKEN_HERE
      threads: 4
"""

with open("/tmp/.dbt/profiles.yml", "w") as f:
    f.write(profiles)

print("profiles.yml written")
```

```python
import subprocess

def run_dbt(command):
    dbt_dir = "/Workspace/Users/singhangad2050@gmail.com/opsintel-copilot/dbt"
    env = {"DBT_PROFILES_DIR": "/tmp/.dbt", "HOME": "/tmp"}
    result = subprocess.run(
        f"cd {dbt_dir} && {command}",
        shell=True,
        capture_output=True,
        text=True,
        env={**__import__('os').environ, **env}
    )
    print(result.stdout)
    if result.returncode != 0:
        print("STDERR:", result.stderr)
        raise Exception(f"Failed: {command}")
```

```python
run_dbt("dbt debug")
```
